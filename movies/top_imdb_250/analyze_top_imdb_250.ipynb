```python
from top_imdb_250.TopIMDb250DB import TopIMDb250DataBase
from pprint import pprint
import re
```

```python
db = TopIMDb250DataBase()
```

# List movies from lowest to highest Runtimes

```python
runtime_list = [(e['Runtime'], e['imdbID']) for e in db.getAllTitles()]
```

```python
runtime_list.sort(key=lambda minutes : int(minutes[0].strip(' min')))
```

```python
#runtime_list
```

```python
for _, id in runtime_list:
    t = db.getOneTitle(imdbID=id)
    print('%s : %s' % (t['Runtime'], t['Title']))
```

# List movies according to their Genre

```python
genre_list = [(e['Genre'], e['imdbID']) for e in db.getAllTitles()]
```

```python
#genre_list
```

```python
genre_dict = {}
for t in genre_list:
    for g in re.split(r',\s*', t[0]):
        genre_dict.setdefault(g, [])
        genre_dict[g].append(t[1])
```

```python
#genre_dict
```

```python
for genre, movie_list in genre_dict.items():
    print('%d titles in %s:' % (len(movie_list), genre))
    for movie_id in movie_list:
        t = db.getOneTitle(imdbID=movie_id)
        print(t['Title'], end=' | ')
    print('\n')
```

## List movies according to the Release Year

```python
year_list = [(e['Year'], e['imdbID']) for e in db.getAllTitles()]
```

```python
#year_list
```

```python
year_dict = {}
for y,i in year_list:
    year_dict.setdefault(y, [])
    year_dict[y].append(i)
```

```python
#year_dict
```

```python
for year in sorted(year_dict):
    year_list = year_dict[year]
    print('%d title%s in %s:' % (len(year_list),'s' if len(year_list) > 1 else '', year))
    for movie_id in year_list:
        t = db.getOneTitle(imdbID=movie_id)
        print(t['Title'], end=' | ')
    print('\n')
```

## List movies from in descending order of score where score = imdbRating x imdbVotes

```python
ratings_votes_list = [(e['imdbRating'], e['imdbVotes'], e['imdbID']) for e in db.getAllTitles()]
```

```python
#ratings_votes_list
```

```python
score_list = [ (float(x)*float(y.replace(',','')), z) for x,y,z in ratings_votes_list]
```

```python
#score_list
```

```python
ratings_votes_list.sort(reverse=True)
```

```python
#ratings_votes_list
```

```python
descending_score_list = sorted(score_list, reverse=True)
```

```python
#descending_score_list
```

```python
for x,y in zip(descending_score_list, ratings_votes_list):
    score, score_id = x
    rating, _, rating_id = y
    score_t, rating_t = db.getOneTitle(imdbID=score_id), db.getOneTitle(imdbID=rating_id)
    # print lists side-by-side to compare
    print('%.2f\t%s\t|\t%s\t%s' % (score, score_t['Title'], rating, rating_t['Title']))
```
